## Backtest Information
### Quant Finance Meetup - July 20, 2017
***Short Term Mean Reversion***

* **Basic Methodology:**
    * Rank Stocks by 5-day Return.
    * Buy the 100 weakest stocks.
    * Short the 100 strongest stocks.
* **Market Exposure:** Dollar Neutral
* **Gross Leverage:** 200%
* **Position Weighting:** Equal-Weighted
* **Rebalance:** 1st trading day of every week
* **Backtest Date:** 2003-01-01 to 2017-07-01
* **Modified Backtest 5 Date:** 2009-01-01 to 2014-01-01
