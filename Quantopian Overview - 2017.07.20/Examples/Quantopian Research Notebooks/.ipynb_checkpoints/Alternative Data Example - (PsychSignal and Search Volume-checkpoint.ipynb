```python
spy_sent_df = odo(spy.sort('asof_date'), pd.DataFrame)
print type(spy_sent_df)
```

```python
# Count the number of days each year that there is a reading
spy_sent_df.groupby(spy_sent_df.timestamp.map(lambda x: x.year)).count()['bull_minus_bear']
```

```python
# Clean the Data by filling in missing dates for plotting and averaging purposes
spy_sent_df.set_index('timestamp', inplace=True, drop=False)
spy_sent_df = spy_sent_df.groupby([spy_sent_df.index.date, 'sid','source']).sum()
spy_sent_df.reset_index(inplace=True)
spy_sent_df.set_index('level_0', inplace=True)
spy_sent_df.index.name='timestamp_date'
spy_sent_df.index=spy_sent_df.index.to_datetime()

# Fill in missing dates with zero NAN
first_date = spy_sent_df.index[0]
end_date = spy_sent_df.index[-1]
spy_sent_df = spy_sent_df.reindex(index=pd.DatetimeIndex(freq='D', start=first_date, end=end_date))

# Fill bull_minus_bear missing values with 0 (This is what we will average and plot)
spy_sent_df['bull_minus_bear'] = spy_sent_df['bull_minus_bear'].fillna(0)

spy_sent_df.tail(3)
```

### Quick Exploratory Performance Analysis

Typically, we would use pipeline to pull the data in and then evaluate the factor on a number of stocks. However, it may be a good idea to "play around" with a small portion of the data (in this case, 1 stock) to get a sense for any idiosyncracies in the data. Finally, it keeps this example simple for demonstration purposes.

```python
def plot_sentiment_data(df):
    # Plot Bull Minus Bear Messages
    ax1 = plt.axes()
    df.plot(y='bull_minus_bear', ax=ax1, label='Bull - Bear Intensity', color='lightblue')
    df.bull_minus_bear.rolling(window=30).mean(). \
        plot(color='green', label='Monthly Avg.: Bull - Bear', ax=ax1)
    plt.ylabel('Bull Minus Bear Message Intensity')
    ax1.axhline(0, color='k')
    # Plot Monthly Average of Total Scanned Messages
    ax2 = plt.twinx(plt.gca())
    df['total_scanned_messages'].rolling(window=30).mean() \
        .plot(color='r', label='Monthly Avg. of Total Scanned Msgs', ax=ax2)
    ax2.grid(False)
    ax2.set_ylabel('Total Scanned Messages')

    plt.title('SPY Stocktwits Sentiment Analysis', fontweight='bold', fontsize=14)
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    plt.legend(lines1+lines2, labels1+labels2, loc='upper center')
    return ax1, ax2

plot_sentiment_data(spy_sent_df);
```

```python
# Let's zoom in
ax1, ax2 = plot_sentiment_data(spy_sent_df.loc['2012':,:])
ax1.set_ylim([-0.5,0.5])
```

```python
def calc_forward_returns(prices, period, dropna=False):
    fwd_returns = prices.pct_change(period).shift(-period)
    if dropna:
        return fwd_returns[:-period]
    return fwd_returns
```

```python
spy_prices = get_pricing('spy', start_date='2012-1-1', 
                         end_date=spy_sent_df.index[-1], fields='open_price')
```

```python
spy_forward_returns = calc_forward_returns(spy_prices, 5, dropna=True)
spy_forward_returns_a = spy_forward_returns.reindex(index=pd.DatetimeIndex(start='2012-1-1', 
                            end=spy_sent_df.index[-1], freq='D', tz='UTC'),
                            method='nearest')
```

```python
ax1 = plt.axes()
ax2 = plt.twinx(ax1)
spy_sent_df['bull_minus_bear'].loc['2012':].rolling(window=30).mean().plot(color='g', ax=ax1)
spy_forward_returns_a.plot(ax=ax2, color='r', label='SPY Forward Returns')
ax2.axhline(0, color='k', linewidth=3)
ax1.grid(False)
ax2.grid(False)
plt.legend()
```

```python
def calc_multiple_fwd_returns(prices, periods=[1,5,10,21]):
    fwd_returns = pd.DataFrame()
    for period in periods:
        fwd_returns[period] = calc_forward_returns(prices, period)
    return fwd_returns
```

```python
PERIODS = [1,5,10,21]
PERIODS = [5,10,21, 65]
spy_fwd_returns = calc_multiple_fwd_returns(spy_prices, periods=PERIODS)
spy_fwd_returns.index.tz=None
spy_fwd_returns.head()
```

```python
merged_data = spy_sent_df.merge(spy_fwd_returns, how='inner', 
                  left_on=spy_sent_df.index.date, 
                  right_on=spy_fwd_returns.index.date)
merged_data.set_index('key_0', inplace=True)
merged_data.index.name ='timestamp'
merged_data.head()
```

```python
import seaborn as sns
fig, ax = plt.subplots(nrows=2, ncols=2)

grid_coords=[(x,y) for x in [0,1] for y in [0,1]]

for period,(x,y) in zip(PERIODS, grid_coords):
    sns.regplot(merged_data['bull_minus_bear'], merged_data[period], ax=ax[x,y])
    ax[x,y].set(title=period)
    
fig.suptitle('Bull - Bear vs. Forward Returns (2010-2014)', fontweight='bold', fontsize=14)
fig.tight_layout()
```

### The example below uses search volume data from Google Trends of global stock market indices.

The idea for this example came from the following paper: <br/>
Tao Chen (2017): Investor Attention and Global Stock Returns, Journal of
Behavioral Finance, <u>DOI:10.1080/15427560.2017.1331235</u>





```python
search_terms = local_csv('country_search_terms.csv')
search_terms.head()
```

```python
country_search_volume = local_csv('country_search_volume_2017.07.02.csv', date_column='date')
country_search_volume
country_search_volume.head()
```

```python
symbol_list = pd.Series(country_search_volume.columns)
mapper = search_terms.set_index('Search Term').Country
country_search_volume.columns = symbol_list.map(mapper)
```

```python
country_search_volume.tail()
```

```python
country_search_volume['Argentina'].plot()
plt.title("Argentina Search Volume (2004-2017)\nNormalized to 100", fontsize=14)
plt.ylabel("Search Volume (% of Max Search Volume)");
```

```python
import seaborn as sns
```

```python
countries_to_plot = ['Argentina', 'Australia', 'Brazil', 'Canada', 'Chile', 'China', 'Colombia', 
                     'France', 'Germany', 'Greece', 'Hong Kong', 'India','Israel', 'Italy', 'Japan',
                    'Mexico', 'Peru', 'Portugal', 'Russia', 'Saudi Arabia', 'South Africa', 'South Korea',
                    'Spain', 'Turkey', 'United Kingdom', 'United States']
```

```python
correlation_matrix = country_search_volume[countries_to_plot].corr()
correlation_matrix.head()
```

```python
mask = np.zeros_like(correlation_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
np.fill_diagonal(mask, False)
sns.heatmap(correlation_matrix, mask=mask, vmin=-1, vmax=1, cmap='RdBu', annot=True, fmt='0.1f',
           annot_kws={'fontsize':10});
plt.title("Correlation of Search Volume by Country", fontweight='bold', fontsize=14);
```
